# Exponential growth

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

## Numerical model

```python
# parameters
mu = 1.
Po = 1.
dt = 0.1

# time aray
t = np.arange(0,5,dt)

#initialize population array
P = np.nan*t
P[0] = Po

# loop through and calculate
for ii in np.arange(len(t)-1):
    P[ii+1] = P[ii] + mu*P[ii]*dt
```

## Analytical model

```python
Pt = Po*np.exp(mu*t)
```

```python
plt.figure()
plt.plot(t,P,'o')
plt.plot(t,Pt,'-')
```

```python
from scipy.integrate import odeint
```

```python
def dPdt(P,t,mu):
    return mu*P

# order of arguments:
# next value of dependent variable
# = (current value of dependent variable, 
# current value of independent variable, usually time or space
# extra parameters)

P2 = odeint(dPdt,Po,t,args=(mu,))
# next value of dependent variable
#  = (function governing change of dependent variable, 
# initial condition (can be a vector), 
# independent variable, 
# extra parameters )
```

```python
plt.figure()
plt.plot(t,P2,'o')
plt.plot(t,Pt,'-')
```
